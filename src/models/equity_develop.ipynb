```python
tickers_name = pd.read_csv('tickers.csv')
tickers_name
```

```python
equities = pd.DataFrame(columns=['Ticker', 'Name', 'Sector', 'Industry', 'Last Price'])
```

```python
for ticker in tickers_name['Ticker']:
    try:
        stock = yf.Ticker(ticker)
        name = stock.info.get('longName')
        sector = stock.info.get('sector')
        industry = stock.info.get('industry')
        last_price = stock.fast_info['lastPrice']
        
        equities = equities.append({
            'Ticker': ticker,
            'Name': name,
            'Sector': sector,
            'Industry': industry,
            'Last Price': last_price
        }, ignore_index=True)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
```

```python
equities
```

```python
equities = equities.dropna(how='any').reset_index(drop=True)
equities
```
