```python
def portfolio_generator(positions, equities_df):
    portfolio_data = []
    for _ in range(positions):
        ticker = random.choice(equities_df['Ticker'])
        number_of_shares = random.randint(10000, 50000)
        equity = equities_df[equities_df['Ticker'] == ticker].iloc[0]
        
        portfolio_data.append({
            'Ticker': ticker,
            'Name': equity['Name'],
            'Sector': equity['Sector'],
            'Industry': equity['Industry'],
            'Number of shares': number_of_shares,
            'Price': equity['Last Price'],
            'Position Value': number_of_shares * equity['Last Price']
        })
    
    return pd.DataFrame(portfolio_data)
        
```

```python
portfolio = portfolio_generator(1000, equities)
```

```python
portfolio
```

```python
portfolio = portfolio.groupby('Ticker', as_index=False).agg({
    'Name': 'first',
    'Sector': 'first',
    'Industry': 'first',
    'Number of shares': 'sum',
    'Position Value': 'sum'
})

portfolio['Price'] = portfolio['Position Value'] / portfolio['Number of shares']
portfolio = portfolio[['Ticker', 'Name', 'Sector', 'Industry', 'Number of shares', 'Price', 'Position Value']]
```
