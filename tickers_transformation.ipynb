```python
import pandas as pd
```

```python
nyse_tickers = pd.read_csv('nyse_tickers.csv')
nyse_tickers_names = nyse_tickers.drop(["Name", "Sector"], axis = 1)
nyse_tickers_names

```

```python
nasdaq_tickers = pd.read_csv('nasdaq_tickers.csv')
nasdaq_tickers_names = nasdaq_tickers.drop(["Name", "Sector"], axis = 1)
nasdaq_tickers_names
```

```python
joint_tickers = pd.concat([nyse_tickers_names, nasdaq_tickers_names], ignore_index=True)
joint_tickers.to_csv("tickers.csv", index=False)
```
